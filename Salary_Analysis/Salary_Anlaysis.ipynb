### Which job has the highest salaries? Illustrate with bar graph.

```python
DF = df.dropna()
```

```python
job_salary = DF.groupby("job_title")["salary"].mean()
max_salary_job = job_salary.idxmax()
max_salary = job_salary.max()
print(f"\n Job with highest salary:{max_salary_job}")
print(f"\n salary of this job: {max_salary}")
```

```python
plt.figure(figsize=(12,9))
x_jobtitle = job_salary.index
y_salary = job_salary.values
plt.bar(x_jobtitle, y_salary, width  = 0.8)
plt.title("Highest paying jobs chart")
plt.xticks(rotation = 90, fontsize = 6)
plt.ylabel('salary')
plt.tight_layout()
plt.show()
```

### Write a python program to find out salaries based on experience level. Illustrate it through bar graph.

```python
NewDF = DF[["job_title","experience_level","salary"]]
```

```python
group_df = NewDF.groupby(["job_title", "experience_level"])["salary"].agg(mean_salary=("mean"))


data = group_df.unstack().plot(kind="bar", stacked=True, width=0.5)

fig = plt.gcf()
fig.set_size_inches(19, 10)
plt.title("Mean Salary by Job Title and Experience Level")
plt.xlabel("Job title")
plt.ylabel("Mean Salary")
plt.legend(title="Experience Level", fontsize = '20')
plt.xticks(rotation=90, fontsize=10)


min_salary = group_df.min().min()  
max_salary = group_df.max().max()  
step = 108


plt.tight_layout()
plt.show();
```
